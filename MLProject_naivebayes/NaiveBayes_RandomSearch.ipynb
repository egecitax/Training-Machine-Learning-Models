```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report, confusion_matrix

dataset = pd.read_csv('normalized_data.csv')
dataset = dataset.drop(["Unnamed: 0"], axis = 1)

x = dataset.drop(["Outcome"], axis = 1)
y = dataset.Outcome.values

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)

param_dist = {
    'priors': [[0.5, 0.5], [0.3, 0.7], [0.8, 0.2]],
    'var_smoothing': [1e-9, 1e-8, 1e-7]
}
nb = GaussianNB()
rs = RandomizedSearchCV(estimator=nb, param_distributions=param_dist, n_iter=5, cv=5, random_state=42)
rs.fit(x_train,y_train)

best_model = rs.best_estimator_
accuracy = best_model.score(x_test, y_test)
print("Accuracy on test set:", accuracy)


y_train_pred = best_model.predict(x_train)
y_test_pred = best_model.predict(x_test)


print("Best hyperparameters:", rs.best_params_,"\n")
print(classification_report(y_train,y_train_pred))
print("\n",classification_report(y_test,y_test_pred))
```
